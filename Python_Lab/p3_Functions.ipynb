```python
# Functions in python
# a, b are parameters of the function
def add(a,b):
    s=a+b
    return s


def product(c=100,d=100):    # by deafult c and d are 100
    p = c*d
    return p

def square(x):
    return x*x
```

```python
print(add)
p=add
print("The sum of given no. is:", p(10,20))

print(f"Type of the add is:{type(add)} and type of p is:{type(p)}")
# 5 - Arguments
result = square(5)
print("The square of given no. is:", result)
```

```python
#Types of arguments
# 1 - Positional arguments:- Values are passed in the same order as parameters are defined.(order matters)
r = product(5,3)
print("The product of given no. is:", r)
# 2 - Keyword arguments(Named Arguments):- Arguments are passed using parameter names, so order doesn’t matter.
r = product(d=5, c=3)
print("The product of given no. is:", r)
#Note: Positional argument cannot appear after keyword argument.(Keyword argument must be at last place)

# 3 - Optional/Default arguments:- Parameters have default values if no value is provided.
r= product()
print("The product of given no. is:", r)
```

```python
# Returning multiple values from a function
def add_product(m =5,n=3):
    s= m+n
    p=m*n
    return s,p

# Tuple packing\
t=add_product()
# Tuple unpacking
sum,prod= t
print(f"Type of t is:{type(t)}")
sum,prod= t
print(f"Sum of two numbers is: {sum}")
print(f"and Product of two numbers is: {prod}")

# normal way
sum,prod= add_product(10,3)
print(f"Sum of two numbers is: {sum}")
print(f"and Product of two numbers is: {prod}")
```

```python
#4 - Variable length Argument - "args"
# *args is used to pass a variable number of non-keyword arguments to a function.

def fun(*args):
    print("Vriable length arguments")
    print(f"Type of args is: {type(args)}")
    t= args
    sum = 0
    for number in t:
        sum+=number
        print(f"Sum of given numbers is: {sum}")

fun(10,20,30)


#5 - Variable keyword argument - "kwargs"
# **kwargs is used to pass a variable number of keyword arguments to a function.

def fun_two(**kwargs):
    print("Keyword arguments")
    print(f"Type of args is:{type(kwargs)}")

    for key,value in kwargs.items():
        print(f"{key}----{value}")

fun_two(Name="Pratik",age="22",dept="IT")    
```

```python
# Lambda function
# Anonymous function- Nameless function
#syntax: lambda input-parameter: expression
s= lambda p,q:p+q
print(f"Sum of two numbers is: {s(10,20)}")
sq = lambda num:num*num
print(f"Square of given number is: {sq(5)}")

greater = lambda a,b: a if a>b else b
print(f"Greater number is: {greater(10,20)}")

gratest_of_three = lambda p,q,r: p if p>q and p>r else (q if q>r else r)
print(f"Gratest of three numbers is: {gratest_of_three(10,200,30)}")
```

```python
#traditional way to differenctiate even no. fro one list to another list
def check_even(number):
    if number%2==0:
        return True
    else:
        return False
    
lst_numbers = [10,20,33,45,45,60,80,34,99]

def list_even(lst):
    lst = []
    for i in range(0,len(lst_numbers)):
        if check_even(lst_numbers[i]):
            lst.append(lst_numbers[i])

    print(f"list of Even numbers are: {lst}")    

list_even(lst_numbers)        
```

```python
filtered_list = list(filter(check_even,lst_numbers))
print(f"Filtered list of even numbers: {filtered_list}")
```

```python
filtered_list = list(filter(lambda num: num%2==0,lst_numbers))
print(f"Filtered list of even numbers: {filtered_list}")

map_list = list(map(lambda num: num*num,filtered_list))
print(f"List of squared even numbers: {map_list}")

from functools import reduce
sum_numbers = reduce(lambda sum,num :sum+num,map_list)
print(f"Sum of all number in map_list is:{sum_numbers}")
```

```python
# Goal: Apply a discount to a list of prices.
# The Task: A store is offering a 25% discount on all items. Given a list of original prices,
# Calculated the new discounted prices.

# The Data: prices = [1200,500,3400,80,150]
# traditional way
def discount(price):
    return price - (price*25/100)
prices = [1200,500,3400,80,150]
discounted_prices = []
for i in prices:
    discounted_prices.append(discount(i))
print(f"Discounted prices are: {discounted_prices}")
```

```python
#solve by using map and lambda function
prices = [1200,500,3400,80,150]
discounted_prices = list(map(lambda price: price - (price*25/100), prices))
print(f"Discounted prices are: {discounted_prices}")
```
