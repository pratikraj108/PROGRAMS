```python
class Student:
    college_name = "SKIT"  #Static variable
    def __init__(self,firstName,lastName):     #constructor (maggic method)
        #self is used to create instance variable
        self.fname= firstName      #fname & lname is instance variable
        self.lname= lastName
        print("Id of self is:",id(self))

    #instance method
    def sayhello(self):
        return f"Hello from {self.fname} {self.lname}"
    
    @staticmethod   #decorator
    def display():
        print(f"College name is: {Student.college_name}")
```

```python
s1 = Student("Pratik", "Raj")
s1.sayhello()
#print("Id of self is:",id(s1))
Student.display()
Student.college_name = "Swami Keshvanand Institute of Technology, M&G, Jaipur"
s2 = Student("Rahul", "Sharma")
s1.display()
s2.display()
```

```python
class DemoSorting:
    def __init__(self,numbers):
        self.numbers = numbers

    def bubble_sort(self):
        n = len(self.numbers)
        for i in range(n):
            for j in range(n-1-i):
                if self.numbers[j] > self.numbers[j+1]:
                    self.numbers[j],self.numbers[j+1] = self.numbers[j+1],self.numbers[j]
        return self.numbers
    
    def selection_sort(self):
        n= len(self.numbers)
        for i in range(n):
            pos =i 
            for j in range(i+1,n):
                if self.numbers[pos]>self.numbers[j]:
                    pos =j 
                self.numbers[i],self.numbers[pos] = self.numbers[pos],self.numbers[i]
        return self.numbers  
    
    def merge_sort(self):
        def merge(left, right):
            merged = []
            i = j = 0

            while i < len(left) and j < len(right):
                if left[i] <= right[j]:
                    merged.append(left[i])
                    i += 1
                else:
                    merged.append(right[j])
                    j += 1

            merged.extend(left[i:])
            merged.extend(right[j:])
            return merged

        def divide(arr):
            if len(arr) <= 1:
                return arr

            mid = len(arr) // 2
            left_part = divide(arr[:mid])
            right_part = divide(arr[mid:])
            return merge(left_part, right_part)

        self.numbers = divide(self.numbers)
        return self.numbers
    
    def insertion_sort(self):
        n = len(self.numbers)
        for i in range(1,n):
            key = self.numbers[i]
            j = i-1
            while j >=0 and key < self.numbers[j]:
                self.numbers[j+1] = self.numbers[j]
                j = j - 1
            self.numbers[j+1] = key
        return self.numbers
    
    def linear_search(self, target):
        for i in range(len(self.numbers)):
            if self.numbers[i] == target:
                return i
        return -1 
    
    def binary_search(self, target):
        arr = sorted(self.numbers)
        low = 0
        high = len(arr) - 1

        while low <= high:
            mid = (low + high) // 2
            if arr[mid] == target:
                return mid
            elif arr[mid] < target:
                low = mid + 1
            else:
                high = mid - 1
        return -1  
```

```python
list_numbers = [10,5,20,25,13,18,45]
ob = DemoSorting(list_numbers)
print("List before sorting is:",list_numbers)
print("List after sorting is:",ob.bubble_sort())
print("List after selection sort is:",ob.selection_sort())
print("List after merge sort is:",ob.merge_sort())
print("List after selection sort:",ob.selection_sort())

result = ob.linear_search(20)
print("Linear Search (target=20): ",f"found at index {result}" if result != -1 else "Not found")

result = ob.binary_search(99)
print("Binary search  (target=99)   :", f"Found at index {result}" if result != -1 else "Not Found")
```
