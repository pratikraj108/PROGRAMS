```python
#Tuple - Immutable - ()
    #tuple packing
t= ("Pratik Raj",22,"SKIT,Jaipur")
print(f"Type of t is:{type(t)}")

#tuple unpacking
name, age, address = ("Pratik Raj",22,"SKIT,Jaipur")
print(f"{name}--{age}--{address}")

#tuple indexing 

print(f"1st member of tuple:{t[0]}...Second mmber is:{t[1]}..Third member is:{t[2]}")
print(f"Pratik Raj" in t)

#iteration over tuple
numbers_tuple = (10,20,30,40,50)
sum = 0
max = 0
for number in numbers_tuple:
    sum = sum + number
    if number > max:
        max = number
print(f"Sum of all numbers is: {sum}")
print(f"maximum among all number:{max}")

#tuple slicing
weekly_temperatures = (30,32,31,29,28,27,26)
weekend_temperatures = weekly_temperatures[5:7]
print(f"Weekend temp are: temp are:{weekend_temperatures}")

weekend_temperatures=weekly_temperatures[0:7:2]
print(f"All week temperature are:{weekend_temperatures}")

weekend_temperatures=weekly_temperatures[0:7:1]
print(f"All week temperature are:{weekend_temperatures}")

weekend_temperatures=weekly_temperatures[::-1]
print(f"All week temperature are in reverse:{weekend_temperatures}")

count_30=weekly_temperatures.count(30)

print(f"Count of 30:{count_30}")
```
