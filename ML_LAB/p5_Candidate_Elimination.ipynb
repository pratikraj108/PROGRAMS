```python
import numpy as np
import pandas as pd

data = pd.read_csv('play_tennis_dataset.csv')
concepts = np.array(data.iloc[:, :-1])
target = np.array(data.iloc[:, -1])
def candidate_elimination(concepts, target):
   # Initialize S and G
   specific_h = concepts[0].copy()
   general_h = [["?" for _ in range(len(specific_h))] for _ in range(len(specific_h))]
   for i, instance in enumerate(concepts):
       if target[i] == "Yes": # Positive example
           for x in range(len(specific_h)):
               if instance[x] != specific_h[x]:
                   specific_h[x] = "?"
           for g in general_h:
               if not all(g[x] == "?" or g[x] == instance[x] for x in range(len(instance))):
                   general_h.remove(g)
       elif target[i] == "No": # Negative example
           for g in general_h:
               for x in range(len(instance)):
                   if specific_h[x] != "?" and specific_h[x] != instance[x]:
                       g[x] = specific_h[x]
                   else:
                       g[x] = "?"
   # Remove redundant hypotheses from G
   general_h = [g for g in general_h if g != ["?"] * len(specific_h)]
   return specific_h, general_h
# Run the algorithm
s_final, g_final = candidate_elimination(concepts, target)
# Output the version space
print("Version space by Candidate Elimination Algorithm:")
print(f"S: {s_final}")
print(f"G: {g_final}")
```
